```python
from cc3dtools.PostProcess import SpacePlot
from cc3dtools.PostProcess import PostProcess
from cc3dtools.GenomeCompare import GenomeCompare
from cc3dtools.Lineage import MultiLineage
from math import pi
```

```python
sp = SpacePlot(start_file='./CLI1/start_cells_Mon_Jun_29_12:55:14_2015.csv', end_file='./CLI1/finish_cells_Mon_Jun_29_12:55:14_2015.csv')
gc = GenomeCompare.from_gen_file('./CLI1/genomes_Mon_Jun_29_12:55:14_2015.csv')
pp = PostProcess(end_file='./CLI1/finish_cells_Mon_Jun_29_12:55:14_2015.csv', gc=gc)
```

```python
cluster_search_results, plot_stack = pp.cluster_search(250,0,0, theta=pi/2, step_size=50, steps=12, cluster_size=25, return_plot_stack=True)
```

```python
sp.plot_all(plot_stack=plot_stack)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%pylab inline
x = np.linspace(1, 50, 50)
for search_result in cluster_search_results:
    PostProcess.plot_frequency_graph(search_result[1], title='distance:'+str(search_result[0]) , plot_stack=[[x , 3000/x, 'r']])
```

```python
plt.plot(x, 1000/x, 'r')
plt.show()
```
