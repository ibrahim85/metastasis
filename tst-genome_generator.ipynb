```python
from cc3dtools.Genome import Genome
```

```python
genomes = []
for x in range(200):
    genomes.append( Genome ( mutation_rate = 178, name = x ) )
```

```python
for g in genomes:
    for x in range(100):
        g.mutate()
```

```python
from collections import Counter
for g in genomes:
    print Counter(g.get_mutated_loci()).most_common(5)
```

```python
replicated_genomes = []
for i,g in enumerate(genomes):
    if i % 2 == 0:
        replicated_genomes.append(g.replicate())
```
