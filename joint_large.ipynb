# Joint Cluster Analysis
This iPython notebook details the tool developed to discover the joint distribution of mutation numbers between two clusters of cells taken from a tumor. The main tool to be used is `PostProcess`.

We first import the necessary modules to deal with this:

````python
%pylab inline
from cc3dtools.PostProcess import SpacePlot
from cc3dtools.PostProcess import PostProcess
from cc3dtools.GenomeCompare import GenomeCompare
from cc3dtools.Lineage import MultiLineage
from math import pi
````

````python
sp = SpacePlot(start_file='./CLI1/start_cells_Mon_Jun_29_12:55:14_2015.csv', end_file='./CLI1/finish_cells_Mon_Jun_29_12:55:14_2015.csv', projection='2d')
gc = GenomeCompare.from_gen_file('./CLI1/genomes_Mon_Jun_29_12:55:14_2015.csv')
pp = PostProcess(end_file='./CLI1/finish_cells_Mon_Jun_29_12:55:14_2015.csv', gc=gc)
````

## Caution
There are two colors of cells just to be able to differentiate a spectral pattern, the apex of the conic is where the cells started growing from. The blue cells also have a higher mutation rate in their genomes compared to the blue cells. However, there is no selective advantage for the blue cells over the red

````python
sp.plot_all()
````

# RANDOM SAMPLLING!
To first demonstrate what happens with the tool, we randomly pick two clusters of 400 cells from our tumor. And computer their marginals. This represents the number of mutations shared by the proportion of the population

````python
import numpy as np
clusters = [ list( np.random.choice(pp.cells_available, 400) ) , list( np.random.choice(pp.cells_available, 400) ) ]
joint = pp.frequency_analyze_ND( clusters )
marginal_1 = pp.frequency_analyze( clusters[0] )
marginal_2 = pp.frequency_analyze( clusters[1] )
````

We first plot the marginals. It looks convicing as most numbers of mutations are shared by a few number of cells (i.e most mutations are private). And only a few of the mutations are shared amongst 20+ cells in the individual clusters.

````python
PostProcess.plot_frequency_graph(marginal_1)
PostProcess.plot_frequency_graph(marginal_2)
````

We can now plot a 2D version between the clusters. As we can see most mutations are shared by a low number of cells in both clusters. The close the spot is to the center line, the more simillar the clusters are. So we obtain that most of the cluster is simillar with a few of pertubations around it.

````python
PostProcess.plot_2D_frequency(joint, xlim=[0,100], ylim=[0, 100])
````

## Validation by trivial case (both clusters are the same)
When both clusters are the same, we expect there to be a straight line through the plot

````python
# obtain two clusters from our tumor
# at (x,y) = (350, 175) and (x,y) = (200, 175)
cluster_1 = map(lambda e: e['id'], pp.nearest( 350 , 175 , 0 , radius = 25) )
cluster_2 = map(lambda e: e['id'], pp.nearest( 200, 175, 0 , radius = 25 ) )


# if you supply the same cluster twice, you expect to see a straight line through the matrix)
joint = pp.frequency_analyze_ND( [ cluster_1 , cluster_1 ] )
PostProcess.plot_2D_frequency( joint )
````

# REAL SAMPLING!
We now do some real sampling in between clusters from different regions of the tumor.

````python
# obtain two clusters
# at (x,y) = (350, 175) and (x,y) = (200, 175)
cluster_1 = map(lambda e: e['id'], pp.nearest( 350 , 175 , 0 , radius = 25) )
cluster_2 = map(lambda e: e['id'], pp.nearest( 200, 175, 0 , radius = 25 ) )
````

````python
# check if they are big enough
len(cluster_1), len(cluster_2)
````

````python
# clusters seem to be spacially close but genetically diverged
joint = pp.frequency_analyze_ND( [ cluster_1 , cluster_2 ] )
PostProcess.plot_2D_frequency(joint)
````

````python
# obtain two clusters, this time closer together but 10 lattice points apart
# it seems that they are more simillar. (could be due to overlap in some cells)
cluster_1 = map(lambda e: e['id'], pp.nearest( 285 , 15 , 0 , radius = 25) )
cluster_2 = map(lambda e: e['id'], pp.nearest( 295, 15, 0 , radius = 25 ) )
# check if they are big enough
print len(cluster_1), len(cluster_2)


joint = pp.frequency_analyze_ND( [ cluster_1 , cluster_2 ] )
PostProcess.plot_2D_frequency(joint)
````

# Multiple Cluster Analysis (y-axis)
What we will attempt to do is traverse the tumor from the opposite ends of the tumor along the y direction and plot at each distance how the frequency density changes

````python
# generate some clusters going accross the tumor in the y direction
clusters, plt_stack = pp.cluster_return(250, 25, 0, theta = pi/2, step_size = 50, steps = 10, cluster_size = 25, return_plot_stack = True)
````

````python
# plot it to see if we nailed the parameters
sp.plot_all( plot_stack = plt_stack )
````

````python
for i in range(len(clusters)/2):
    print clusters[i][0], clusters[-i-1][0]
    data = pp.frequency_analyze_ND( [ clusters[i][1] , clusters[-i-1][1] ] )
    PostProcess.plot_2D_frequency(data , title='Clusters at y = '+str(clusters[i][0])+' and y = '+str(clusters[-1-i][0]) )
    
````
