# Comparison of CSC and MutationResponse Limbo Models

## Methodology
- Run Limbo model to grow normal cells for 50,000MCS
- Once 50,000MCS completes, we take this *Template* and we run a model on top of it, either Cancer Stem Cell (CSC) or the MutationResponse models.
- Here are the investigations of some early results.

## Visual Inspection
### 1st Order Model
- Growth occurs only in cancer cells.

```python
Image(filename='./simulation_out/1stOrder/Limbo1stOrder_Wednesday_12082015_105359/pipe_out_Wednesday_12082015_134527/spatial_plot.png')
```

### CSCs Model
- Growth occurs only in the Cancer Stem Cells (Green cells)
- Tumor cells have a limited division potential
- CSCs can loose their potential and even give rise to more CSCs with a small probability, if they do not they give rise to regular tumor cells

```python
Image(filename='./simulation_out/CSCs/LimboCSC_Thursday_13082015_165319/pipe_out_Thursday_13082015_194141/spatial_plot.png')
```

### MutationResponse Model
- Cancer cells grow in proportion to the number of mutations in a short range of their genome upto a maximum.

```python
Image(filename='./simulation_out/MutationResponse/LimboMutationResponse_Thursday_13082015_135541/pipe_out_Thursday_13082015_152532/spatial_plot.png')
```
